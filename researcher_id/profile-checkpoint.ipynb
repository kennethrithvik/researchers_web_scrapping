```python
import pandas as pd 
from urllib.request import urlopen
import json 
from bs4 import BeautifulSoup as bs
import timeit
import json
import requests
import traceback
import re
```

```python
ids = pd.read_csv("ids.csv")
list_ids = ids["Id"].values.tolist()
list_ids = list_ids[101:200]
```

```python
start_time = timeit.default_timer()
with open('two.json', 'w') as f:



 
    for research_id in list_ids:
###################################################################
        try:

            url_profile = "http://www.researcherid.com/rid/" + research_id
            print(url_profile)
            open_url_profile = requests.get(url_profile)
            profile_soup = bs(open_url_profile.content , "html.parser")
            
            
        

####################################################################

        
            if profile_soup.find('input',{'name':'researcher.id'}).attrs["value"]:
                spl_id= profile_soup.find('input',{'name':'researcher.id'}).attrs["value"]
                #spl_id = profile_soup.find("input" , attrs={'name' : 'researcher.id' }).attrs["value"]
            else :spl_id="null"
                
                
            if profile_soup.find('div',{'class':'profileName'}):
                name= profile_soup.find('div',{'class':'profileName'}).text
                name = " ".join(name.split())
            else :name="null"
            
          
            if profile_soup.find('td',string="Subject:"):
                container = profile_soup.find('td',string="Subject:").parent
                subject = container.find("td",{"class":"profileDataCells"}).text
                subject = " ".join(subject.split()).split(';')
            else :subject="null" 
                
            if profile_soup.find('td',string="Keywords:"):
                container = profile_soup.find('td',string="Keywords:").parent
                Keywords = container.find("td",{"class":"profileDataCells"}).text
                Keywords = " ".join(Keywords.split()).split(';')
            else :Keywords="null" 
                
            if profile_soup.find('td',string="Publons:"):
                container = profile_soup.find('td',string="Publons:").parent
                Publons = container.find("td",{"class":"profileDataCells"}).text
                Publons = " ".join(Publons.split()).split(';')
            else :Publons="null"    
                
                
            if profile_soup.find('td',string="ORCID:"):
                container = profile_soup.find('td',string="ORCID:").parent
                ORCID = container.find("td",{"class":"profileDataCells"}).text
                ORCID = " ".join(ORCID.split()).split(';')
            else :ORCID="null" 
                
            if profile_soup.find('td',string="Primary Institution:"):
                container = profile_soup.find('td',string="Primary Institution:").parent
                Primary_Institution = container.find("td",{"class":"profileDataCells"}).text
                Primary_Institution = " ".join(Primary_Institution.split()).split(';')
            else :Primary_Institution="null" 
                    
            if profile_soup.find('td',string="Role:"):
                container = profile_soup.find('td',string="Role:").parent
                Role = container.find("td",{"class":"profileDataCells"}).text
                Role = " ".join(Role.split()).split(';')
            else :Role="null" 
                
                
            if profile_soup.find('label',{'id':'current_metadata_total'}):
                pub_no= profile_soup.find('label',{'id':'current_metadata_total'}).text
            else: pub_no= 0
#########################################################################################                
            Publication=[]
            Patent=[]
            if pub_no is not 0 :
                url_profile = "http://www.researcherid.com/WorkspacePaging.action?&current.metadata.size=800&current.metadata.pageSizeUpdated=true&researcherId="+spl_id
                open_url_profile = requests.get(url_profile)
                profile_soup = bs(open_url_profile.content , "html.parser")
                reg = re.compile('summary_data.*')
                list_pubs = profile_soup.find_all("td",{"class":reg})
                
                i=0
                if list_pubs[0].find("input",{"name":"artifacts[0].timesCited"}):
                    for pubs in list_pubs:
                        pub_obj ={}
                        tit= "artifacts["+str(i)+"].itemTitle"
                        aur="artifacts["+str(i)+"].listOfAuthorsAsDescription"
                        dat="artifacts["+str(i)+"].date"
                        cit="artifacts["+str(i)+"].timesCited"
                        scr = "artifacts["+str(i)+"].sourceTitle"


                        pub_obj={'Title':pubs.find("input",{"name":tit}).attrs["value"] if pubs.find("input",{"name":tit}) else "null" , 
                             'Authors': pubs.find("input",{"name":aur}).attrs["value"].split(';') if pubs.find("input",{"name":aur}) else "null" ,
                             'Date': pubs.find("input",{"name":dat}).attrs["value"] if pubs.find("input",{"name":dat}) else "null",
                             'Cited': pubs.find("input",{"name":cit}).attrs["value"] if pubs.find("input",{"name":cit}) else "null"
                                }
                        Publication.append(pub_obj)

                        i=i+1
                else:
                    for pubs in list_pubs:
                        pat_obj ={}
                        tit= "artifacts["+str(i)+"].itemTitle"
                        ass= "artifacts["+str(i)+"].patentAssigneeNames"
                        inv= "artifacts["+str(i)+"].listOfAuthorsAsDescription"




                        pat_obj={'Title':pubs.find("input",{"name":tit}).attrs["value"] if pubs.find("input",{"name":tit}) else "null" , 
                                 'Assignee': pubs.find("input",{"name":ass}).attrs["value"].split(';') if pubs.find("input",{"name":ass}) else "null",
                                 'Inventors':pubs.find("input",{"name":inv}).attrs["value"].split(';') if pubs.find("input",{"name":inv}) else "null"
                                }
                        Patent.append(pat_obj)

                        i=i+1

                        
                

                
                
            
     
            
            
                
            
                
               
                
                
######################################################################
            user={'spl_id':spl_id, 
                  'name': name, 
                  'r_id':research_id,
                  'subject':subject,
                  'Keywords':Keywords,
                  'Publons':Publons,
                  'ORCID':ORCID,
                  'Primary_Institution':Primary_Institution,
                  'Role':Role,
                  'pub_no':pub_no,
                  'Patent':Patent,
                  'Publication':Publication
                  
                  
                 }
                
######################################################################        
        except Exception as ex:
            print (ex)
            user={'r_id':research_id}
                
            
            
            
######################################################################
        
        try:
            json.dump(user, f, ensure_ascii=True)
            f.write("\n")
        except er:
            print(er)


###################################################################
f.close()
elapsed = timeit.default_timer() - start_time
print(elapsed)
traceback.print_exc()
```
