```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
filename = "hello.txt"
file = open(filename, "w")
```

```python
html= requests.get('https://www.tripadvisor.com.sg/Attraction_Review-g294264-d2439664-Reviews-Universal_Studios_Singapore-Sentosa_Island.html#REVIEWS')
soup=bs(html.text,'html.parser')
reviewDivs = soup.find_all("div", {"class": "review-container"})
html2= requests.get('https://www.tripadvisor.com.sg/Attraction_Review-g294264-d2439664-Reviews-or10-Universal_Studios_Singapore-Sentosa_Island.html')
soup2=bs(html2.text,'html.parser')
reviewDivs2=soup2.find_all("div", {"class": "review-container"})
reviewDivs=reviewDivs[:]+reviewDivs2[:]
for review in reviewDivs:
    try:
        if review.div.div.div.find('span',{"class": "userLocation"}) :
            location=review.div.div.div.find('span',{"class": "userLocation"}).text+",null"
        else:
            location="null,null"
        userName=review.div.div.div.find('span',{"class": "scrname"}).text
        reviewText=review.div.div.div.find('p',{"class": "partial_entry"}).text
        reviewTitle=review.div.div.div.find('span',{"class": "noQuotes"}).text
        rating=review.div.div.div.find('div',{"class": "rating"}).span['class'][1]
        rating=int(rating.split('_')[1])/10.0
        user=review.div.div.div.find('div',{"class": "memberBadgingNoText"})
        postCount=user.find_all('span',{"class":"badgetext"})[0].text
        usefulVotes=user.find_all('span',{"class":"badgetext"})[1].text
        print "Review Title = "+reviewTitle    
        print "Review Text = "+reviewText
        print "User Name = "+userName
        print "Location :- { city: "+location.split(',')[0]+" , country: "\
                            +location.split(',')[1]+" }"
        print "Rating = ",rating
        #file.write("Rating = "+str(rating))
        print "No. of posts by user = ",postCount
        print "No. of useful votes by user = ",usefulVotes
        print "\n\n"
    except Exception as ex:
        print ex
file.close()
```
